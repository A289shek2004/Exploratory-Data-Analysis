# Merge and Concate

```python
v=pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]})
vv=pd.DataFrame({'A':[1,2,3,4],'C':[21,22,23,24]})
```

```python
pd.merge(v,vv,on='A')
```

```python
v=pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]})
vv=pd.DataFrame({'A':[1,2,3,5],'C':[21,22,23,24]})
```

```python
pd.merge(v,vv)
```

```python
pd.merge(v,vv,how='outer')
```

```python
pd.merge(v,vv,how='outer',indicator=True)
```

```python
pd.merge(v,vv,how='left')
```

```python
pd.merge(v,vv,how='right')
```

```python
pd.merge(v,vv,on='A')
```

```python
v=pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]})
vv=pd.DataFrame({'A':[1,2,3,5],'B':[21,22,23,24]})
```

```python
pd.merge(v,vv,left_index=True,right_index=True)
```

```python
pd.merge(v,vv,left_index=True,right_index=True,suffixes=('_name','_python'))
```

```python
#Concat
s1=pd.Series([1,2,3,4])
s2=pd.Series([11,12,13,14])
pd.concat([s1,s2])
```

```python
v=pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]})
vv=pd.DataFrame({'A':[1,2,3,4],'C':[21,22,23,24]})
pd.concat([v,vv])
```

```python
pd.concat([v,vv],axis=1)
```

```python
pd.concat([v,vv],axis=1,join='inner')# or outer
```

```python
pd.concat([v,vv],axis=1,keys=['s1','s2'])
```

```python
v1=pd.DataFrame({'D':[1,2,3,4]})
v2=pd.DataFrame({'A':[1,2,3,4],'C':[21,22,23,24]})
pd.concat([v1,v2])
```
