#  Data Visualization with Matplotlib


```python
import matplotlib.pyplot  as plt
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
c=['red','blue','green','yellow','pink']
plt.bar(x,y,color=c)
plt.show()
```

# Bar Plot

```python
import matplotlib.pyplot as plt
x=['python','c++','java','c']
y=[85,70,80,65]
col=['red','blue','green','yellow','pink']
plt.title("Simple Bar Plot")
plt.xlabel('Languages',fontsize=10)
plt.ylabel("Popularity")
plt.bar(x,y,color=col,width=0.5,edgecolor='black',alpha=0.2)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=['python','c++','java','c']
y=[85,70,80,65]
col=['red','blue','green','yellow','pink']
plt.title("Simple Bar Plot")
plt.xlabel('Languages',fontsize=10)
plt.ylabel("Popularity")
plt.bar(x,y,color=col,label='abhishek')
plt.legend()
plt.show()
```
